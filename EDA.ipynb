```python
rb = pd.read_csv('RB_14to24.csv')
```

```python
rb.head()
```

```python
# # Sort by player and season first
# rb = rb.sort_values(['gsis_id', 'season']).reset_index(drop=True)

# # Group by gsis_id only, cumsum, then shift to exclude current season
# rb['career_carries'] = rb.groupby('gsis_id')['carries'].cumsum().shift(1).fillna(0)

rb = rb.sort_values(['gsis_id', 'season']).reset_index(drop=True)

# Create cumulative sum first
rb['cumsum_carries'] = rb.groupby('gsis_id')['carries'].cumsum()

# Then shift within each group
rb['career_carries'] = rb.groupby('gsis_id')['cumsum_carries'].shift(1).fillna(0)
```

```python
rb.head()
```

```python
plt.scatter(rb['career_carries'], rb['player_points.total'])
plt.xlabel('Career Carries')
plt.ylabel('Season Score')
plt.show()
```

```python
coaches_avg = rb.groupby('OC')['player_points.total'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(range(len(coaches_avg)), coaches_avg.values)
plt.xticks(range(len(coaches_avg)), coaches_avg.index, rotation=90)
plt.xlabel('Head Coach')
plt.ylabel('Average Fantasy Points')
plt.title('Average RB Fantasy Points by Head Coach')
plt.tight_layout()
plt.show()
```
